```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
final = pd.DataFrame()
for j in range(1, 333):
    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebkit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 safari/537.36'}
    webpage = requests.get(url, headers=headers).text
    oup = BeautifulSoup(webpage, 'lxml')
    company = soup.find_all('div', class_ = 'company-content-wrapper')
    
    name = []
    ratings = []
    reviews = []
    ctype = []
    hq = []
    old = []
    employee = []
    des = []
    
    for i in company:
        name.append(i.find('h2').text.strip())
        ratings.append(i.find('p', class_='rating').text.strip())
        reviews.append(i.find('a', class_='review-count').text.strip())
        ctype.append(i.find_all('p', class_='infoEntity')[0].text.strip())
        hq.append(i.find_all('p', class_='infoEntity')[1].text.strip())
        old.append(i.find_all('p', class_='infoEntity')[2].text.strip())
        employee.append(i.find_all('p', class_='infoEntity')[3].text.strip())
        des.append(i.find_all('p', class_='description')[0].text.strip())

    d = {'name':name, 'ratings':ratings, 'reviews' : reviews, 'type':ctype, 'head_quarter':hq, 'old': old, 'employees' : employee, 'description':des}
    
    df = pd.DataFrame(d)
    final = final.append(df, ignore_index=True)
   
```
